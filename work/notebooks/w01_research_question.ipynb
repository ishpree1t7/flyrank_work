## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
from IPython.display import Markdown, display

display(Markdown("""
## Research Question

### Question
Which content pages should be prioritized for review based on observable search and engagement signals?

### Decision
Determine which pages should be reviewed first.

### Action
A content manager reviews the highest-ranked pages and decides whether to refresh, rewrite, expand, monitor, or leave them unchanged.

### Unit of Analysis
One content page.

### Output
A ranked review queue with refresh scores and reason codes.

### Cost of a Wrong Recommendation

False Positive:
Editorial time is wasted reviewing pages that did not require action.

False Negative:
Important declining pages are missed, causing lost traffic opportunities.

### Why ML?

Machine learning combines many observable signals such as impressions, CTR, freshness, engagement, and position to prioritize pages better than simple fixed rules while still supporting human decisions.
"""))
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)

print("Unique Pages:",
      df["content_id"].nunique())

print("Average Impressions (90d):",
      round(df["impressions_90d"].mean(),2))

print("Average Sessions (90d):",
      round(df["sessions_90d"].mean(),2))

declining = (df["trend_direction"]=="down").mean()*100

print("Declining Pages:",
      round(declining,2),"%")
```

## 4. Careful words: what I can and can't claim
